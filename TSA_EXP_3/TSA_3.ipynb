```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Generate synthetic air quality dataset
dates = pd.date_range(start="2020-01-01", periods=100, freq='D')
np.random.seed(42)
aqi_values = np.linspace(50, 150, 100) + np.random.normal(0, 10, 100)

df = pd.DataFrame({'Date': dates, 'AQI Value': aqi_values})
df.set_index('Date', inplace=True)

# Perform the Augmented Dickey-Fuller (ADF) Test
def adf_test(series):
    result = adfuller(series.dropna())
    print("Augmented Dickey-Fuller Test:")
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"   {key}: {value}")
    if result[1] <= 0.05:
        print("The time series is stationary.")
    else:
        print("The time series is NOT stationary.")

# Check stationarity of original AQI data
print("Checking stationarity for Original AQI Data")
adf_test(df['AQI Value'])

# Plot the original AQI data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['AQI Value'], label='AQI Value', color='b', linewidth=2)
plt.title("Air Quality Time Series Data", fontsize=14, fontweight='bold')
plt.xlabel("Date")
plt.ylabel("AQI Value")
plt.legend()
plt.grid()
plt.show()
```
