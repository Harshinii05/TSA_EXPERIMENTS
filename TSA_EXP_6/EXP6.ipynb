```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic air quality dataset
dates = pd.date_range(start="2020-01-01", periods=100, freq='D')
np.random.seed(42)
aqi_values = np.linspace(50, 150, 100) + np.random.normal(0, 10, 100)

df = pd.DataFrame({'Date': dates, 'AQI Value': aqi_values})
df.set_index('Date', inplace=True)

# Apply moving average smoothing (window size = 5)
df['Smoothed'] = df['AQI Value'].rolling(window=5, center=True).mean()

# Plot only smoothed data
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Smoothed'], label='Smoothed Data (Moving Average)', color='r', linewidth=2)
plt.title("Air Quality Index - Moving Average Smoothing")
plt.xlabel("Date")
plt.ylabel("AQI Value")
plt.legend()
plt.show()
```
